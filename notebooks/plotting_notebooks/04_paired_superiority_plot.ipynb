# PairSuperiorityPlot

Quick usage example (wide format).


```python
import pandas as pd
from calvin_utils.plotting_utils.pair_superiority_plot import PairSuperiorityPlot

# Load your CSV (paired columns)
csv_path = '/path/to/your.csv'
df = pd.read_csv(csv_path)

plotter = PairSuperiorityPlot(
    stat_array_1=df['stat_a'],
    stat_array_2=df['stat_b'],
    model1_name='A',
    model2_name='B',
    stat='Metric',
    out_dir=None,
    method='bootstrap',
)
plotter.draw()
```
