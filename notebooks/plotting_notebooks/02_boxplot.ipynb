Quick usage example (wide format).


```python
import pandas as pd
csv_path = '/Users/cu135/Partners HealthCare Dropbox/Calvin Howard/studies/memory_dbs/results/vta_centroids_comparison/overall/bilateral/centroids_paired.csv'
df = pd.read_csv(csv_path)
```

Unpaired comparison

```python
from calvin_utils.plotting_utils.simple_box_plot import SimpleBoxPlotWrapper
plotter = SimpleBoxPlotWrapper(df)
plotter.plot(
    columns=['x_a', 'y_a', 'z_a', 'l2_a'],
    group_labels=['X', 'Y', 'Z', 'L2'],
    dataset_name='VTA Centroids',
    xlabel='Coordinate Direction',
    ylabel='Distance (mm)',
    figsize=(11,5),
    out_dir=None
)
```

Pairwise Comparison

```python
from calvin_utils.plotting_utils.simple_box_plot import SimpleBoxPlotWrapper

plotter = SimpleBoxPlotWrapper(df)
plotter.plot(
    columns=[('x_a', 'x_b'), ('y_a','y_b'), ('z_a','z_b'), ('l2_a', 'l2_b')],
    group_labels=['X', 'Y', 'Z', 'L2'],
    pair_names=['Clinician', 'Optimizer'],
    dataset_name='VTA Centroids',
    xlabel='Coordinate Direction',
    ylabel='Distance (mm)',
    figsize=(11,5),
    out_dir='/Users/cu135/Partners HealthCare Dropbox/Calvin Howard/studies/memory_dbs/results/vta_centroids_comparison/overall/bilateral/norm'
)
```
